```python
def rotations(combo):
  position = 50
  zero_count = 0
  for pos in combo:
    print(pos)
    if pos.startswith('L'):
      position = (position - int(pos.split('L')[1])) % 100
    else:
      position = (position + int(pos.split('R')[1])) % 100
    
    # print(position)
    if position == 0:
      zero_count += 1

  return zero_count
```

```python
rotations(input_combo)
```

```python
def rotations_part2(combo):
    position = 50
    zero_count = 0

    for move in combo:
        direction = move[0]
        diff = int(move[1:])

        if direction == 'R':
            # distance to next 0
            offset = 100 - position
            if diff >= offset:
                zero_count += 1 + (diff - offset) // 100
            position = (position + diff) % 100

        else:  # 'L'
            offset = position
            if offset == 0:
                offset = 100
            if diff >= offset:
                zero_count += 1 + (diff - offset) // 100
            position = (position - diff) % 100

    return zero_count
```

```python
rotations_part2(combo)
```
