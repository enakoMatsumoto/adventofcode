```python
total = 0
n = len(nums)
count = 0
temp = []
for i in range(len(nums[0])):
  try:
    temp.append(int(''.join(nums[j][i] for j in range(n))))
  except:
    if ops[count] == '*':
      total += np.prod(temp)
    else:
      total += np.sum(temp)
    count += 1
    temp = []

if ops[count] == '*':
  total += np.prod(temp)
else:
  total += np.sum(temp)
```
