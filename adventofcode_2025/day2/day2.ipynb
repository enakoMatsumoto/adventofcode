```python
from typing import List
import math

# updated for part2 such that it is generalized for all cases

def find_invalid_id_sum_part2WRONG(range_list: List[str]): # THIS IS WORNG BECAUSE IT DOUBLE COUNTS INSTANCES LIKE 1111 FOR BOTH BLOCKS OF 1 AND BLOCKS OF 2
    start, end = range_list
    L, R = int(start), int(end)
    start_digit_len = len(start)
    end_digit_len = len(end)

    total = 0

    for digit_len in range(start_digit_len, end_digit_len + 1):

        for block_len in range(1, digit_len // 2 + 1):
            if digit_len % block_len != 0:
                continue

            repeat_count = digit_len // block_len
            if repeat_count < 2: # “some sequence of digits repeated at least twice”
                continue

            # multiplier like 1001, 1001001, 1010101, etc
            multiplier = (10 ** digit_len - 1) // (10 ** block_len - 1)

            # block bounds (no leading zeros)
            first_valid_block = 10 ** (block_len - 1)
            last_valid_block = 10 ** block_len - 1

            # digit-length boundary handling
            if digit_len == start_digit_len:
                first_valid_block = max(
                    first_valid_block,
                    math.ceil(L / multiplier)
                )

            if digit_len == end_digit_len:
                last_valid_block = min(
                    last_valid_block,
                    math.floor(R / multiplier)
                )

            if first_valid_block > last_valid_block:
                continue

            # arithmetic sum: (a + b) * n / 2, multiplied by multiplier
            count = last_valid_block - first_valid_block + 1
            total += (
                (first_valid_block + last_valid_block) * count // 2
            ) * multiplier

    return total
```

```python
from typing import List
import math

def find_invalid_id_sum_part2(range_list: List[str]):
    start, end = range_list
    L, R = int(start), int(end)
    start_digit_len = len(start)
    end_digit_len = len(end)

    total = 0
    seen_numbers = set()  # Track numbers we've already counted

    for digit_len in range(start_digit_len, end_digit_len + 1):

        for block_len in range(1, digit_len // 2 + 1):
            if digit_len % block_len != 0:
                continue

            repeat_count = digit_len // block_len
            if repeat_count < 2:
                continue

            # multiplier like 1001, 1001001, 1010101, etc
            multiplier = (10 ** digit_len - 1) // (10 ** block_len - 1)

            # block bounds (no leading zeros)
            first_valid_block = 10 ** (block_len - 1)
            last_valid_block = 10 ** block_len - 1

            # digit-length boundary handling
            if digit_len == start_digit_len:
                first_valid_block = max(
                    first_valid_block,
                    math.ceil(L / multiplier)
                )

            if digit_len == end_digit_len:
                last_valid_block = min(
                    last_valid_block,
                    math.floor(R / multiplier)
                )

            if first_valid_block > last_valid_block:
                continue

            # generate numbers from first_valid_block to last_valid_block
            for block in range(first_valid_block, last_valid_block + 1):
                num = block * multiplier
                if num < L or num > R:
                    continue
                if num in seen_numbers:
                    continue  # avoid double counting
                seen_numbers.add(num)
                total += num

    return total
```
