```python
from typing import List

def find_invalid_id_sum(range_list: List[str]):
  start, end = range_list
  start_digit_len = len(start)
  end_digit_len = len(end)

  total = 0

  if start_digit_len == end_digit_len:
    if start_digit_len % 2 == 1:
      return 0
    half_digit_len = int(start_digit_len / 2)
    start_first_half, start_second_half = int(start[:half_digit_len]), int(start[half_digit_len:])
    end_first_half, end_second_half = int(end[:half_digit_len]), int(end[half_digit_len:])

    # first_valid_half = max(start_first_half, start_second_half)
    first_valid_half = start_first_half
    if start_first_half < start_second_half:
      first_valid_half += 1
    last_valid_half = end_first_half
    # while last_valid_half < end_first_half:
    #   last_valid_half += 1

    if not (last_valid_half == end_first_half and last_valid_half <= end_second_half):
      last_valid_half -= 1

    total = int((first_valid_half + last_valid_half)*(last_valid_half-first_valid_half+1) / 2) * (10 ** half_digit_len + 1)
      
    return total

  for digit_len in range(start_digit_len, end_digit_len + 1):
    if digit_len % 2 == 1:
      continue

    half_digit_len = int(digit_len / 2)
    power_num = 10 ** half_digit_len
    if digit_len == start_digit_len:
      start_first_half, start_second_half = int(start[:half_digit_len]), int(start[half_digit_len:])
      # first_valid_half = max(start_first_half, start_second_half)
      first_valid_half = start_first_half
      if start_first_half < start_second_half:
        first_valid_half += 1
      last_valid_half = power_num - 1
    elif digit_len == end_digit_len:
      end_first_half, end_second_half = int(end[:half_digit_len]), int(end[half_digit_len:])
      first_valid_half = power_num // 10
      if end_second_half != 0:
        last_valid_half = min(end_first_half, end_second_half)
      else:
        if end_first_half == first_valid_half:
          continue # no valid entry if end_first_half == 100 for example where let's say range was 90,000 to 100,000. then there are no valid example of ids we want to find
        last_valid_half = end_first_half - 1 # if end_first_half == 200 because the range is to 200,000, then we should accept up to 199,199
    else:
      first_valid_half = power_num // 10
      last_valid_half = power_num - 1

    total += int((first_valid_half + last_valid_half)*(last_valid_half-first_valid_half+1) / 2) * (power_num + 1)

  return total

```

```python
ex_input = '11-22,95-115,998-1012,1188511880-1188511890,222220-222224,1698522-1698528,446443-446449,38593856-38593862'
```

```python
sum([find_invalid_id_sum(range_str.split('-')) for range_str in ex_input.split(',')])
```
