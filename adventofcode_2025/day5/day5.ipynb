```python
def get_input(file):
  with open(file, 'r') as file:
    input = file.read().split('\n\n')

  fresh_ranges = [list(map(int, r.split('-'))) for r in input[0].split('\n')]
  available_list = list(map(int, input[1].split('\n')))

  return fresh_ranges, available_list
```

```python
ex_fresh_ranges, ex_available_list = get_input('example.txt')
```

```python
def merge_ranges(ranges):
    sorted_ranges = sorted(ranges, key= lambda x: x[0])
    merged = []
    for start, end in sorted_ranges:
        if not merged or merged[-1][1] < start:
            merged.append([start, end])
        else:
            merged[-1][1] = max(merged[-1][1], end)
    return merged
```

```python
def fresh_ingredients(fresh_ranges, available_list):
  cleaned_fresh_ranges = merge_ranges(fresh_ranges)
  sorted_available_list = sorted(available_list)

  # both ranges and available ids are sorted, so can do two pointer search
  range_pointer = avail_pointer = 0
  inside_range, outside_range = [], []

  while avail_pointer < len(sorted_available_list):
      if range_pointer == len(cleaned_fresh_ranges) or sorted_available_list[avail_pointer] < cleaned_fresh_ranges[range_pointer][0]:
          outside_range.append(sorted_available_list[avail_pointer])
          avail_pointer += 1
      elif sorted_available_list[avail_pointer] > cleaned_fresh_ranges[range_pointer][1]:
          range_pointer += 1
      else:
          inside_range.append(sorted_available_list[avail_pointer])
          avail_pointer += 1

  return len(inside_range)
```

```python
fresh_ingredients(ex_fresh_ranges, ex_available_list)
```

```python
fresh_ranges, available_list = get_input('input.txt')
```

```python
fresh_ingredients(fresh_ranges, available_list)
```
